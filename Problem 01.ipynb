## Biểu đồ

```python
features = ["Loại cửa hàng", "Số sao trên ShopeeFood", "Giá thấp nhất", "Giá cao nhất", "Số sao trên Foody", 
            "Phí dịch vụ", "Số lượt đánh giá trên Foody", "Giờ mở cửa", "Giờ đóng cửa", "Thành phố", "Quận", 
            "Seeding", "Thành công"]
```

```python
def QuanAn(df, name):
    df_QuanAn = df.loc[df["Chi tiết cửa hàng"] == name]
    df_QuanAn = df_QuanAn[features]
    return df_QuanAn
```

```python
def pie_chart(df, name):
    df_QuanAn = QuanAn(df, name)
    pie = df_QuanAn["Thành công"].value_counts()

    plt.figure(figsize=(10,6))
    plt.pie(pie, labels=["Không thành công", "Thành công"] , autopct="%1.2f%%", colors=["tab:blue", "tab:orange"],
           wedgeprops={"edgecolor":"black"}, textprops={'fontsize': 12})
    plt.title(f"Biểu đồ thể hiện tỉ lệ số cửa hàng \nthành công của cửa hàng {name}".format(name), loc="center")
    plt.legend()

    plt.show()
```

```python
def scatter_plot(df, name):
    df_QuanAn = QuanAn(df, name)
    x = df_QuanAn.loc[df_QuanAn["Thành công"] == True]
    y = df_QuanAn.loc[df_QuanAn["Thành công"] == False]

    plt.figure(figsize=(12,8))
    plt.scatter(y["Giá thấp nhất"], y["Giá cao nhất"], label="Không thành công", s=50)
    plt.scatter(x["Giá thấp nhất"], x["Giá cao nhất"], c="tab:orange", label="Thành công", s=50)
    
    plt.xlabel("Giá thấp nhất")
    plt.ylabel("Giá cao nhất")
    plt.legend(loc="lower right")
    plt.title(f"Biểu đồ thể hiện giá thấp nhất và giá cao nhất của cửa hàng {name}".format(name), loc="center")

    plt.show()
```

```python
def barh_chart_1(df, name):
    df_QuanAn = QuanAn(df, name)
    a = df_QuanAn.loc[df_QuanAn["Thành công"] == True]["Loại cửa hàng"].value_counts().sort_values(ascending=False).values
    b = df_QuanAn.loc[df_QuanAn["Thành công"] == False]["Loại cửa hàng"].value_counts().sort_values(ascending=False).values[:a.shape[0]]
    c = df_QuanAn.loc[df_QuanAn["Thành công"] == True]["Loại cửa hàng"].value_counts().sort_values(ascending=False).index

    x_axis = np.arange(len(a))
    width = 0.4

    fig, ax = plt.subplots(figsize=(12, 8))
    ax.barh(x_axis, a, width, color="tab:orange", label='Thành công', edgecolor="black")
    ax.barh(x_axis + width, b, width, color="tab:blue", label='Không \nthành công', edgecolor="black")

    ax.set(yticks=x_axis + width, yticklabels=c, ylim=[2*width - 1, len(x_axis)])
    ax.legend(loc='upper right')
    plt.title(f"Biểu đồ thể hiện số lượng nhãn của cửa hàng {name}".format(name), loc="center")

    plt.show()
```

```python
def line_chart(df, name):
    df_QuanAn = QuanAn(df, name)
    
    def location(city):    
        x = df_QuanAn.loc[df_QuanAn["Thành công"] == True]
        a = x.loc[x["Thành phố"] == city]["Quận"].value_counts().sort_index()
        b = df_QuanAn.loc[df_QuanAn["Thành phố"] == city]["Quận"].value_counts().sort_index()
        c = (a / b).sort_values()
        return c
    
    plt.figure(figsize=(24, 16))
    
    plt.subplot(2,1,1)
    plt.plot(location(" TP. HCM"))
    plt.xlabel("Quận")
    plt.ylabel("Tỉ lệ thành công")
    plt.title(f"Biểu đồ thể hiện tỉ lệ thành công cửa hàng {name} tại thành phố Hồ Chí Minh".format(name), loc="center")
    
    plt.subplot(2,1,2)
    plt.plot(location(" Hà Nội"))
    plt.xlabel("Quận")
    plt.ylabel("Tỉ lệ thành công")
    plt.title(f"Biểu đồ thể hiện tỉ lệ thành công cửa hàng {name} tại thành phố Hà Nội".format(name), loc="center")
    
    plt.savefig("1.png")

    plt.show()
```

```python
def barh_chart_2(df, name):
    df_QuanAn = QuanAn(df, "Trà Sữa")

    x = df_QuanAn.loc[df_QuanAn["Thành công"] == True]
    y = df_QuanAn.loc[df_QuanAn["Thành công"] == False]

    opening_hour_1 = []
    opening_hour_2 = []

    for i in x.index:
        for j in range(x["Giờ mở cửa"][i], x["Giờ đóng cửa"][i] + 1):
            opening_hour_1.append(j)
    for i in y.index:
        for j in range(y["Giờ mở cửa"][i], y["Giờ đóng cửa"][i] + 1):
            opening_hour_2.append(j)

    opening_hour_1 = pd.Series(opening_hour_1)
    opening_hour_2 = pd.Series(opening_hour_2)

    opening_hour_1 = opening_hour_1.value_counts()
    opening_hour_2 = opening_hour_2.value_counts()

    width = 0.4
    fig, ax = plt.subplots(figsize=(12, 8))
    ax.barh(opening_hour_1.index, opening_hour_1.values, width, color="tab:orange", 
            label='Thành công', edgecolor="black")
    ax.barh(opening_hour_2.index + width, opening_hour_2.values, width, color="tab:blue", 
            label='Không \nthành công', edgecolor="black")

    ax.legend(loc='lower right')
    plt.xlabel("Số lượng cửa hàng đang mở cửa")
    plt.ylabel("Thời gian đang mở cửa")
    plt.title(f"Biểu đồ thể hiện số lượng cửa hàng {name} đang mở cửa tại các thời điểm trong ngày".format(name), 
              loc="center")
    plt.show()
```
